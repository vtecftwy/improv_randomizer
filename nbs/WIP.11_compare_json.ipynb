```python
import json
from pathlib import Path
```

```python
p2config = Path('../config').resolve()
p2config
```

```python
paths = 'ai2/ai2-games-all.json ai2/ai2-games-may-2024.json ai2/ai2-games.json'
paths = paths + ' ai2-show/ai2-games-2024-01-20.json ai2-show/ai2-games-2024-05-18-z15.json'

game_jsons = {}
for path in paths.split(' '):
    p = p2config / path
    if not p.is_file():
        print(p)
    else:
        game_jsons[path] = json.loads(p.read_text())
```

List of all game names

```python
game_names = set()

for games in game_jsons.values():
    for game in games.values():
        game_names.add(game['name'])
len(game_names)
```

```python
game_names
```

```python
def print_values(d, keys=None):
    if keys is None: keys = 'nbr_players nbr_audience prompt exclude host_include host_exclude description tips category'.split(' ')
    for key in keys:
        print(f"{key}: {d.get(key, '-')}")
```

```python
for game_name in sorted(game_names):
    for p, games in game_jsons.items():
        for game in games.values():
            if game['name'] == game_name:
                print(p)
                print_values(game)
    print()
```
